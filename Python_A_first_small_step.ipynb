# Data Types and Variables

The basic data types are:
- integer (e.g. 12)
- float (e.g. 3.34)
- compex (e.g. 1+3j)
- boolean (TRUE,FALSE)
- string (e.g. "Hello World")

Variables are names for values and the symbol "=" assigns a value on the right to the name on the left. The variable is created when a value is assigned (no declaration necessary).

Variable names:
- cannot start with a digit
- cannot contain spaces, quotation marks, or other punctuation
- may contain an underscore (typically used to separate words in long variable names)
- Underscores at the start like “__test” have a special meaning so we won’t do that until we understand the convention


Some examples:

```python
intValue = 12
floatValue = 3.34
cValue = complex(1,3)
bValue = True
strValue = "Hello World"

print(intValue)
print(floatValue)
print(cValue)
print(bValue)
print(strValue)
```

# Operations

The basic arithmetic operations are:
<pre>
- Addtion:              +
- Substraction:         -
- Multiplication:       *
- Division:             /
- Exponentiation:       **
</pre>

Some examples:

```python
a = 1
b = 2
c = 3.3
d = complex(1,3)

print(a + b)
print(a - b)
print(b*c)
print(c/b)
print(d**0.5)
```

# Decisions

Python supports the usual comparison operators:
<pre>
- Equals:                       a == b
- Not Equals:                   a != b
- Less than:                    a < b
- Less than or equal to:        a <= b
- Greater than:                 a > b
- Greater than or equal to:     a >= b
</pre>

These operators can be used in several ways, most commonly they are used in "if statements" 


A first simple example <span style="color:red;">**(pay attention to the colon and the indentation)** </span>:

```python
a = 1
b = 2

if a < b:
    print("a is lower than b")
```

The first line of the if statment must end with a colon, and the body must be indented.
- The colon at the end of the first line signals the start of a block of statements.
- Python uses indentation rather than {} or begin/end to show nesting.
- Any consistent indentation is legal, but almost everyone uses four spaces.
- All Python IDEs or also even simple editors are doing the indentation automatically


The elif keyword is Python's way of saying "if the previous conditions were not true, then try this condition" and the else keyword catches anything which isn't caught by the preceding conditions.

To combine conditions the logical operators AND and OR can be used.



Example:

```python
temperature = 0

if temperature > 25.0 and temperature < 60:
    print("The temperature is", temperature, ". Seems to be summer!")
elif temperature < 10.0 and temperature > 0.0:
    print("The temperature is", temperature, ". Seems to be winter but it is not freezing!")
elif temperature <= 0.0:
    print("The temperature is", temperature, ". It is freezing!")
elif temperature == 23.0 or temperature == 25.0:
    print("The temperature is", temperature, ". This is one of my favourite temperatures")
else:
    print("No statement for this temperature")
```

# Loops

- A for loop tells Python to execute some statements once for each value in a list, a character string, or some other collection.
- “for each thing in this group, do these operations”


Example <span style="color:red;">**(again pay attention to the colon and the indentation)** </span>:

```python
for names in ["Eddy", "Philip", "John"]:
    print(names)
```

Use range to iterate over a sequence of numbers
- Use it to simply repeat an action some number of times
- or use it to iterate over a sequence of numbers
- or use a step if it is needed (so it is range(start, stop, step))

Examples:

```python
for i in range(5):
    print("Hello")

print()

for i in range (10,16):
    print(i)

print()

for n in range(17,26,2):
    print(n)
```

If another condition for the end of a loop is needed the while loop can be used
- The condition for the end of the loop can be set on the beginning
- The condition for the end of the loop can be set somewhere within the block

Examples <span style="color:red;">**(again pay attention to the colon and the indentation)** </span>:

```python
i = 0
while i < 6:
    print(i)
    i = i + 1

print()

i = 0
while 1:
    print(i)
    if i > 6:
        break
    i = i + 1

```

# Functions

A function is a block of code which only runs when it is called.

You can pass data, known as parameters, into a function.

A function can return data as a result.

In Python a function is defined using the **def** keyword

Example:

```python
def hello():
    print("Hello! My name is Py.")
    
hello()
```

Information can be passed into functions as arguments.

Arguments are specified after the function name, inside the parentheses. You can add as many arguments as you want, just separate them with a comma.

To let a function return a value, use the **return** statement.

```python
def hello(name, age):
    print("Hello! My name is " + name + "and I am " + str(age) + " years old.")
    return age + 1
    
nextage = hello("John", 42)
print("Next year John will be " + str(nextage) + " years old.")
```

# Lists and Arrays

A list is a data structure that's built into Python and holds a collection of items:
- List items are enclosed in square brackets
- Lists are ordered – i.e. the items in the list appear in a specific order
- You can add or remove items after a list's creation.
- Item duplication is possible, as each element has its own distinct place and can be accessed separately through the index.
- Elements can be of different data types
- Each element in the list has an index



Some examples:

```python
list1 = [1,2,3,4,5]
print(list1)
list2 = ["Graz", 25.1, 42.5]
print(list2)
print(list2[0])
```

An array data structure belongs to the "must-import" category. To use an array in Python, you'll need to import this data structure from the NumPy package or the array module. And that’s the first difference between lists and arrays.
- Arrays need to be declared. Lists don't, since they are built into Python.
- Arrays can store data very compactly and are more efficient for storing large amounts of data.
- Arrays are great for numerical operations; lists cannot directly handle math operations

```python
import numpy as np

array1 = np.array([1,2,3,4])
print(array1)
array2 = array1 / 3
print(array2)
```

**When list and when array?**
- If you need to store a relatively short sequence of items and you don't plan to do any mathematical operations with it, a list is the preferred choice. This data structure will allow you to store an ordered, mutable, and indexed sequence of items without importing any additional modules or packages.
- If you have a very long sequence of items, consider using an array. This structure offers more efficient data storage.
- If you plan to do any numerical operations with your combination of items, use an array. Data analytics and data science rely heavily on (mostly NumPy) arrays.


# Libraries

A library is a collection of utility methods, classes, and modules that your application code can use to perform specific tasks without writing the functionalities from scratch.
Python’s standard library is very extensive, offering a wide range of facilities. The library contains built-in modules (written in C) that provide access to system functionality such as file I/O that would otherwise be inaccessible to Python programmers, as well as modules written in Python that provide standardized solutions for many problems.
In addition to the standard library, there is an active collection of hundreds of thousands of components available to create applications and models in a variety of fields.
Let’s have a look at some of the commonly used libraries:

- **Matplotlib** (for beginners see e.g.: https://www.simplilearn.com/tutorials/python-tutorial/matplotlib): This library is responsible for plotting numerical data. And that’s why it is used in data analysis. It is also an open-source library and plots high-defined figures like pie charts, histograms, scatterplots, graphs, etc.
- **Pandas** (getting started: https://pandas.pydata.org/docs/getting_started/intro_tutorials/): Pandas is an important library for data scientists. It is an open-source machine learning library that provides flexible high-level data structures and a variety of analysis tools. It eases data analysis, data manipulation, and cleaning of data. Pandas support operations like Sorting, Re-indexing, Iteration, Concatenation, Conversion of data, Visualizations, Aggregations, etc.
- **Numpy** (for beginners see e.g.: https://numpy.org/doc/stable/user/absolute_beginners.html): The name “Numpy” stands for “Numerical Python”. It is the commonly used library. It is a popular machine learning library that supports large matrices and multi-dimensional data. It consists of in-built mathematical functions for easy computations. Even libraries like TensorFlow use Numpy internally to perform several operations on tensors. Array Interface is one of the key features of this library.
- **SciPy** (Introduction see e.g.: https://www.w3schools.com/python/scipy/index.php): The name “SciPy” stands for “Scientific Python”. It is an open-source library used for high-level scientific computations. This library is built over an extension of Numpy. It works with Numpy to handle complex computations. While Numpy allows sorting and indexing of array data, the numerical data code is stored in SciPy. It is also widely used by application developers and engineers.
- **math**: This module provides access to the mathematical functions (e.g. trigonometric functions)

A program must import a library module before using it.
- Use import to load a library module into a program’s memory.
- Then refer to things from the module as *module_name.thing_name*.
    - Python uses . to mean “part of”.


Example:

```python
import math

print('pi is', math.pi)
print('cos(pi) is', math.cos(math.pi))
```

- Use *from ... import ...* to load only specific items from a library module.
- Then refer to them directly without library name as prefix.

But pay attantion that you will not see anymore from which library the function is coming, because no information of a library is there.

```python
from math import cos, pi

print('cos(pi) is', cos(pi))
```

- Use *import ... as ...* to give a library a short alias while importing it.
- Then refer to items in the library using that shortened name.

This is the mainly used from to import a library.

```python
import math as m

print('cos(pi) is', m.cos(m.pi))
```

Use *help* to learn about the contents of a library module

```python
help(math)
```

# Reading data

To begin processing data, we need to load it into Python. We can do that using a library called NumPy, which stands for Numerical Python. In general, you should use this library when you want to do fancy things with lots of numbers, especially if you have matrices or arrays. To tell Python that we’d like to start using NumPy, we need to import it and with the command loadtxt we can load a text data file. To store the data, we use a variable called data:

```python
import numpy

data = numpy.loadtxt(fname='WeatherData.txt', delimiter=',', skiprows=1)
print(data)
```

We have to skip the first rows because in the data file there is a header of two lines. First line is a string with the filename and the second line is the head of the columns.
Now we have the data in a numpy array called data and we can work with it.

Again we can use *help* to learn about the features:

```python
help(numpy.loadtxt)
```

Other formats of textfiles are also possible (First file are temperature values of different thermal zones from a bulding simulation, and second file is a monthly water production of a simulated desalination system using solar power):

```python
import numpy as np

data = np.genfromtxt("Temperature.out", skip_header = 2)
print(data)
```

```python
import numpy as np

labels = np.genfromtxt("water.dat", usecols=0, dtype=str, skip_header = 2, max_rows = 12)
data = np.genfromtxt("water.dat", usecols=1, dtype=float, skip_header = 2, max_rows = 12)

print(labels)
print(data)
```

# Making diagrams

If the weather data are loaded we can make a diagram. To do so we have to import the library *matplotlib.pyplot* In this case it is the ambient temperature over the whole year:


```python
import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt(fname='WeatherData.txt', delimiter=',', skiprows=1)

time = data[0:,0]
temperature = data[0:,3]
plt.plot(time, temperature)
plt.xlabel("Time [h]")
plt.ylabel("Temperature [°C]")
plt.grid()
plt.show()
```

Also for the water production data we can have a diagram:

```python
import numpy as np
import matplotlib.pyplot as plt

labels = np.genfromtxt("water.dat", usecols=0, dtype=str, skip_header = 2, max_rows = 12)
data = np.genfromtxt("water.dat", usecols=1, dtype=float, skip_header = 2, max_rows = 12)

fig,ax = plt.subplots()
plt.bar(labels,data)
fig.autofmt_xdate(rotation=45)
plt.ylabel("water production [m³]")
plt.show()
```

Also diagrams for functions are possible:

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,360,500)
y_sin = np.sin(np.radians(x))
y_cos = np.cos(np.radians(x))

fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.set_title("Sinus- and Cosinusfunction")
ax.set_xlabel("angle [degree]")
ax.set_ylabel("function value")
ax.set_xlim(0,360)
ax.set_ylim(-1.1,1.1)
ax.set_xticks(np.arange(0,361,90)) # (min, max, step) increase the maximum by 1 to get the last x label
ax.grid()


ax.plot(x,y_sin, label = "sinus")
ax.plot(x,y_cos, label = "cosinus")
ax.legend()

plt.show()
```

It is also possible to have more then one plot on a page:

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import FormatStrFormatter


x = np.linspace(0,360,500)
y_sin = np.sin(np.radians(x))
y_cos = np.cos(np.radians(x))

fig = plt.figure()
ax = fig.add_subplot(2,2,1)
ax.set_title("Sinus (blue)", fontsize=20)
ax.plot(x,y_sin,'blue')

ax = fig.add_subplot(2,2,2)
ax.set_title("Cosinus (red)", fontsize=20)
ax.plot(x,y_cos,'red')

ax = fig.add_subplot(2,2,3)
ax.set_ylim(-1.5,1.5)
ax.set_yticks(np.arange(-1.5,1.6,0.5)) 
ax.set_title("Cosinus (green)", fontsize=20)
ax.plot(x,y_cos,'green')

ax = fig.add_subplot(2,2,4)
ax.set_ylim(-2.0,2.0)
ax.set_xlim(0,360)
ax.set_xticks(np.arange(0,361,90)) # (min, max, step) increase the maximum by 1 to get the last x label
ax.yaxis.set_major_formatter(FormatStrFormatter('% 1.2f')) # to have two decimal digits at the y-scale
ax.set_title("Sinus (black)", fontsize=20)
ax.grid()
ax.plot(x,y_sin,'black')

fig.tight_layout() # without this line the title of the second subplot crashes into the scale of the first one

plt.show()
```
